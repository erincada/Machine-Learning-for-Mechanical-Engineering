# Initial Dataset - C-MAPSS FD001 
The classical C-MAPSS FD001 dataset (Saxena & Goebel, 2008) is used as an initial benchmark to develop and validate the data preprocessing pipeline, feature engineering steps, and baseline models. Its well-understood structure and extensive use in the literature allow rapid prototyping and fair comparison before extending the analysis to the more realistic 2021 run-to-failure dataset.

```python
import os
import json
import opendatasets as od
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Authenticate using the existing kaggle.json in the 'other' directory
json_path = '../other/kaggle.json'

if os.path.exists(json_path):
    with open(json_path, 'r') as f:
        creds = json.load(f)
    
    # Set environment variables for authentication
    os.environ['KAGGLE_USERNAME'] = creds['username']
    os.environ['KAGGLE_KEY'] = creds['key']
    print("Authentication successf,ul using ../other/kaggle.json")
else:
    print(f"Error: Could not find {json_path}. Please check the file location.")

# Define download directory
data_dir = '../data/raw'

if not os.path.exists(data_dir):
    os.makedirs(data_dir)

# Download the dataset
dataset_url = 'https://www.kaggle.com/datasets/behrad3d/nasa-cmaps'
od.download(dataset_url, data_dir)
```
